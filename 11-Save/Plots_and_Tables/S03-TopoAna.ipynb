```python
from scipy import stats
def get_pulls(counts,pdf):
    pull = (-pdf + counts)
    return pull
```

```python
def Plot(X, y, Title, Color1, Color2, Sample):

    lw=3

    fig,axs=plt.subplots(2,1,figsize=(10,10),gridspec_kw={'height_ratios':[1,0.2]})

    bins = 50
    bin_edges = np.linspace(0.1,0.55,bins)
    Range = [0.1, 0.55]
    perBin = ((Range[1] - Range[0])/bins)*1000
    
#     test_bkg_count_weight=bins/len(X)
    test_bkg_counts,test_bkg_bins = np.histogram(X,bins=bins,range=(0.1,0.55))

    train_bkg_counts,train_bkg_bins,_etc=axs[0].hist(y,color =Color2,
            histtype='step',bins=bins,density=False,range=(0.1,0.55),linewidth=lw,label='Truth-matching')
    axs[0].hist(y,color = Color2,
            histtype='stepfilled',alpha=0.4,bins=bins,density=False,range=(0.1,0.55))
    bin_width=test_bkg_bins[1]-test_bkg_bins[0]
    bin_centers=[el+(bin_width/2) for el in test_bkg_bins[:-1]]

    axs[0].errorbar(bin_centers,test_bkg_counts,
                yerr=np.sqrt(test_bkg_counts),label='TopoAna',color=Color1,
                marker='o',linewidth=lw,ls='')
    axs[0].set_title(Title,loc='left')
    axs[0].set_title(Sample,loc='right')
    axs[0].set_xlim(0.1,0.55)
    axs[0].set_ylim(0)
    axs[0].set_ylabel(r'$Entries/(\; {width:.2f}\;MeV/c^2)$'.format(width = perBin))

    leg=axs[0].legend(loc='upper center')
    leg._legend_box.align = "left"  

    pulls=get_pulls(test_bkg_counts,np.array(train_bkg_counts))
    axs[1].bar(bin_centers,pulls,width=bin_width)
    axs[1].set_xlim(0.1,0.55)
    axs[1].set_ylabel('Residual')
#     axs[1].set_ylim(-5,5)
    axs[1].set_xlabel(r'$\Delta m(D_s^{+} - D^{0})\;[GeV/c^{2}]$')

    return
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute cos(K, ) from 3-momenta
def compute_cos_angle(px1, py1, pz1, px2, py2, pz2):
    p1 = np.stack([px1, py1, pz1], axis=1)
    p2 = np.stack([px2, py2, pz2], axis=1)
    dot = np.einsum('ij,ij->i', p1, p2)
    mag1 = np.linalg.norm(p1, axis=1)
    mag2 = np.linalg.norm(p2, axis=1)
    cos_theta = dot / (mag1 * mag2)
    return cos_theta

# Only consider true electrons
mask = (abs(DataFrames["uds"]["e_mcPDG"]) == 11)

# Get relevant data
dm = DataFrames["uds"].loc[mask, "Ds_diff_D0pi"]
cos_Kpi = compute_cos_angle(
    DataFrames["uds"].loc[mask, "K_px"],
    DataFrames["uds"].loc[mask, "K_py"],
    DataFrames["uds"].loc[mask, "K_pz"],
    DataFrames["uds"].loc[mask, "pi_px"],
    DataFrames["uds"].loc[mask, "pi_py"],
    DataFrames["uds"].loc[mask, "pi_pz"],
)

# Create 2D histogram
plt.hist2d(cos_Kpi, dm, bins=[50, 50], cmap='viridis')
plt.colorbar(label='Entries')

plt.title(r'$uds \; event$', loc='left')
plt.title(r'$\int\mathcal{L}dt\approx\;1444$ fb$^{-1}$', loc='right')
plt.xlabel(r'$\cos\theta(K,\;\pi)$')
plt.ylabel(r'$\Delta m_{\pi}(D_s^+ - D^0)\;[GeV/c^2]$')
plt.show()
```
