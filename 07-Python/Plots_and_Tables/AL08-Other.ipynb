```python
Stacked = False
Density = True
Bins = 50
i = 'Ds_diff_D0pi'
# i = 'Ds_massDifference_0'
Range = [0.1, 0.55]
dM = 0.1
perBin = ((Range[1] - Range[0])/Bins)*1000
print("Width Per Bin: {width:.2f} MeV".format(width = perBin))

labels=[
    r"$D^{0}:421$",
    r"$D_{1}(2420)^{+}:10413$",
    r"$D_{1}(2430)^{0}:20423$",
    r"$D_{2}^{*}(2460)^{0}:425$",
    r"$D_{0}^{*}(2400)^{0}:10421$",
    r"$D_{1}(H)^{+}:20413$",
    r"$D_{0}^{*}(2400)^{+}:10411$",
    r"$D_{2}^{*}(2460)^{+}:415$",
    ]

data=[
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==421) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==10413) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==20423) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==425) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==10421) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==20413) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==10411) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==415) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      DataFrames["ccbar"][(abs(DataFrames["ccbar"]["Ds_mcPDG"])==20413) & (DataFrames["ccbar"]['Ds_gammaveto_M_Correction']>=dM)][i],
      ]

factor = 0.005
# plt.hist(DataFrames["Signal"][(DataFrames["Signal"]['Ds_gammaveto_M_Correction']>=dM)][i], label="Signal", histtype='step', density=Density, bins=Bins, alpha=1, range=Range, weights=factor*np.ones_like(DataFrames["Signal"][(DataFrames["Signal"]['Ds_gammaveto_M_Correction']>=dM)][i]), ls='--', linewidth=1.5)
plt.hist(data, label=labels, density=Density, stacked=Stacked, bins=Bins, alpha=1, histtype='step', linewidth=1.5, range=Range)
# plt.axvspan(Range[0],0.15,color='gray',alpha=0.2)
# plt.axvline(0.15,ls='--',color='gray')

# Title
#--------
plt.title(r'$Other \; Category$', loc = "left")
# Label
#-------
plt.ylabel(r'$Entries/(\; {width:.2f}\;MeV/c^2)$'.format(width = perBin))
plt.xlabel(r'$\Delta m(D_s^{+} - D^{0})\;[GeV/c^{2}]$')
# plt.yscale("log")
# plt.ylim(0, 10000)
plt.legend()
plt.show()
```
